```python
#median
def stimulate_boot(data,sample_size,no_sample,CI):
    median=[]
    for i in range(no_sample):
        sample=np.random.choice(data,size=sample_size)
        median.append(np.median(sample))
    median=np.array(median)
    sort=np.sort(median)
    per=np.percentile(sort,CI)
    return (sort[0],per)
data=np.random.randint(150,190,500)
sort=stimulate_boot(data,30,1000,95)
```

```python
def hypo(class1,class2,no_sample):
    diff=[]
    mean1=np.mean(class1)
    mean2=np.mean(class2)
    total_mean=mean1-mean2
    concate=np.concatenate([class1,class2])
    for i in range(no_sample):
        np.random.shuffle(concate)
        s1=concate[:len(class1)]
        s2=concate[len(class1):len(class1)+len(class2)]
        s=np.mean(s1)-np.mean(s2)
        diff.append(s)
    diff_arr=np.array(diff)
    sort_data=np.sort(diff_arr)
    p_value=(np.sum(sort_data>total_mean)/no_sample)*100
    return p_value
```

```python
def permutation_test(sample1,sample2,no_samples):
    TS=abs(np.mean(sample1)-np.mean(sample2))
    concate=np.concatenate([sample1,sample2])
    new_data=[]
    for i in range(no_samples):
        np.random.shuffle(concate)
        s1=concate[:len(sample1)]
        s2=concate[len(sample1):]
        new_data.append(abs(np.mean(s1)-np.mean(s2)))
    diff_arr=np.array(new_data)
    p_value=np.mean(diff_arr>=TS)
    return p_value
sample1=np.random.randint(50,100,50)
sample2=np.random.randint(50,100,50)
no_samples=1000
permutation_test(sample1,sample2,no_samples)
```

```python
#permutation test
def permutation(sample1,sample2,no_samples):
    TS=abs(np.mean(sample1)-np.mean(sample2))
    diff=[]
    combine=np.concatenate([sample1,sample2])
    for i in range(no_samples):
        np.random.shuffle(combine)
        s1=combine[:len(sample1)]
        s2=combine[len(sample1):]
        diff.append(abs(np.mean(s1)-np.mean(s2)))
    diff_arr=np.array(diff)
    p_value=np.mean(diff_arr>TS)
    return p_value
sample1=np.array([71, 82, 96, 54, 58, 95, 76, 57, 67, 60, 54, 72, 94, 51, 59, 70, 88,59, 83, 89, 51, 63, 55, 88, 62, 51, 85, 85, 63, 61])
sample2=np.array([65, 59, 68, 92, 61, 78, 59, 72, 99, 56, 77, 59, 80, 53, 56, 88, 95,94, 74, 54, 54, 73, 82, 71, 51, 78, 83, 90, 54, 89])
no_samples=1000
permutation(sample1,sample2,no_samples)
```

```python
a=np.arange(1,10)
from scipy.stats import anderson
result=anderson(a,dist='norm')
print("AD Statistic:", result.statistic)
print("Critical Values:", result.critical_values)
print("Significance Levels:", result.significance_level)
if result.statistic>result.critical_values[2]:
    print('H0 is rejected')
else:
    print("Fail to reject H0")
```
